```python
# Initialize MarkerCluster
marker_cluster = MarkerCluster()

# For each launch record, create a Marker and add it to the cluster
for index, record in spacex_df.iterrows():
    # Marker coordinates
    location = [record['Lat'], record['Long']]
    
    # Popup label with site + outcome
    popup_label = f"{record['Launch Site']} | {'Success' if record['class'] == 1 else 'Failure'}"
    
    # Create the marker
    marker = folium.Marker(
        location=location,
        popup=popup_label,
        icon=folium.Icon(color='white', icon_color=record['marker_color'], icon='rocket', prefix='fa')
    )
    
    # Add marker to the cluster
    marker_cluster.add_child(marker)

# Add the cluster to the map
site_map.add_child(marker_cluster)

# Display map
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Example launch site coordinates (KSC LC-39A for demo)
launch_site_lat = 28.573255
launch_site_lon = -80.646895

# Example proximity points (replace with actual MousePosition coordinates you collected)
city_lat, city_lon = 28.5383, -81.3792        # Orlando
rail_lat, rail_lon = 28.5721, -80.5853        # example railway point
highway_lat, highway_lon = 28.5628, -80.5773  # example highway point
coast_lat, coast_lon = 28.56367, -80.57163    # coastline (from before)

# Helper function to add a marker + line for any proximity point
def add_distance_marker(site_lat, site_lon, point_lat, point_lon, map_obj, color, label):
    distance = calculate_distance(site_lat, site_lon, point_lat, point_lon)
    
    # Draw line between site and proximity point
    folium.PolyLine(
        locations=[[site_lat, site_lon], [point_lat, point_lon]],
        color=color
    ).add_to(map_obj)
    
    # Add marker at the proximity point with distance label
    marker = folium.Marker(
        [point_lat, point_lon],
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html=f'<div style="font-size: 12px; color:{color};"><b>{label}: {distance:.2f} KM</b></div>'
        )
    )
    map_obj.add_child(marker)

# Apply for each proximity type
add_distance_marker(launch_site_lat, launch_site_lon, city_lat, city_lon, site_map, "red", "City")
add_distance_marker(launch_site_lat, launch_site_lon, rail_lat, rail_lon, site_map, "green", "Railway")
add_distance_marker(launch_site_lat, launch_site_lon, highway_lat, highway_lon, site_map, "purple", "Highway")
add_distance_marker(launch_site_lat, launch_site_lon, coast_lat, coast_lon, site_map, "blue", "Coastline")

# Show updated map
site_map
```
